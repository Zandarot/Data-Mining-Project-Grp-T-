```python
methods = ["RandomTree", "RandomTree w/ SMOTE", "IBk", "IBk w/ SMOTE"]
```

```python
average_mse = [40.60824557613789, 0.4517878539738055, 23.946031019330526, 0.3959609628365093]

# Plot for Average MSE
plt.figure(figsize=(10, 6))
plt.bar(methods, average_mse, color='orange', edgecolor='black')
for i, val in enumerate(average_mse):
    plt.text(i, val + 0.5, f"{val:.2f}", ha='center', fontsize=10)
plt.title("Average Mean Squared Error", fontsize=14)
plt.xlabel("Method", fontsize=12)
plt.ylabel("Average Mean Square Error", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
average_mae = [4.96496653883968, 0.23121770227570285, 3.8495880049111877, 0.20185909980430528]

# Plot for Average MAE
plt.figure(figsize=(10, 6))
plt.bar(methods, average_mae, color='blue', edgecolor='black')
for i, val in enumerate(average_mae):
    plt.text(i, val + 0.1, f"{val:.2f}", ha='center', fontsize=10)
plt.title("Average Mean Absolute Error", fontsize=14)
plt.xlabel("Method", fontsize=12)
plt.ylabel("Average MAE", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
